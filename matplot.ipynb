```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
n = 20
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
area = np.pi*(15 * np.random.rand(n))**2
plt.scatter(x,y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
# y = mx + c
m = 5
c = 3
y = m*x + 10
plt.scatter(x,y) # scattering
plt.show()
plt.plot(x,y) # linear plotting
```

```python
sec = 'sec-a', 'sec-b', 'sec-c', 'sec-d','sec-e'
strength = (45,54,32,60,39)
differ = (0.1,0,0,0,0)
plt.pie(strength, explode=differ, labels=sec, autopct='%1.2f%%', shadow=True,startangle=90,)
plt.axis('equal')
plt.show()
```

```python
#histogram
mew = 100
sigma = 25
ab = mew+sigma*np.random.randn(1000)
n, bins, patches=plt.hist(ab,50,facecolor='yellow', alpha=0.8)
plt.title('Histogram')
plt.xlabel('smart graphs')
plt.ylabel('probability')
plt.axis([10,180,1,80])
plt.grid(True)
plt.show()
```
